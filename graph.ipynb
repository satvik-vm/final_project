```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib qt
```

```python
def get_data(file):
	df = pd.read_csv(file)
	train_loss = []
	eval_loss = []
	train_accuracy = []
	eval_accuracy = []
	epoch = []
	for i in range(len(df)):
		if df.iloc[i]['Epoch'] % 10 == 0:
			epoch.append(df.iloc[i]['Epoch'])
			train_loss.append(df.iloc[i]['Train Loss'])
			eval_loss.append(df.iloc[i]['Validation Loss'])
			train_accuracy.append(df.iloc[i]['Train Accuracy'])
			eval_accuracy.append(df.iloc[i]['Validation Accuracy'])
	return epoch, train_loss, train_accuracy, eval_loss, eval_accuracy
```

```python
def plot_graph(file, image_title):
	df = pd.read_csv(file)
	train_loss = []
	eval_loss = []
	train_accuracy = []
	eval_accuracy = []
	epoch = []
	for i in range(len(df)):
		if df.iloc[i]['Epoch'] % 10 == 0:
			epoch.append(df.iloc[i]['Epoch'])
			train_loss.append(df.iloc[i]['Train Loss'])
			eval_loss.append(df.iloc[i]['Validation Loss'])
			train_accuracy.append(df.iloc[i]['Train Accuracy'])
			eval_accuracy.append(df.iloc[i]['Validation Accuracy'])

	fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

	axes[0].plot(epoch, eval_loss, color='red', label='Validation Loss')
	axes[0].plot(epoch, train_loss, color='blue', label='Train Loss')
	axes[0].legend()
	axes[0].set_title('Loss')

	axes[1].plot(epoch, train_accuracy, color='red', label='Train Accuracy')
	axes[1].plot(epoch, eval_accuracy, color='blue', label='Validation Accuracy')
	axes[1].legend()
	axes[1].set_title('Accuracy')

	fig.suptitle(image_title)

	return epoch, train_loss, train_accuracy, eval_loss, eval_accuracy
```
